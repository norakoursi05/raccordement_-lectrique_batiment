```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

#Charger le fichier shapefile
gdf_infrastructure = gpd.read_file('data/infrastructures.shp')
df = pd.read_csv("data/reseau_en_arbre.csv")
batiment = gpd.read_file("data/batiments.shp")
```

```python
#Afficher les premières lignes pour explorer les données
print(gdf_infrastructure.columns)
print(gdf_infrastructure.head)
import matplotlib.pyplot as plt

#Créer une nouvelle figure et des axes
fig, axs = plt.subplots(1, 1, figsize=(20, 20))

#Plot des bâtiments
batiment.plot(ax=axs, color='blue', edgecolor='black', label='Bâtiments')

#Filtrer et plot des infrastructures intactes
infra_intactes = gdf_infrastructure[gdf_infrastructure['infra_id'].isin(df[df['infra_type'] == 'infra_intacte']['infra_id'])]
infra_intactes.plot(ax=axs, color='green', edgecolor='black', label='Infrastructures intactes')

#Filtrer et plot des infrastructures détruites
infra_detruites = gdf_infrastructure[gdf_infrastructure['infra_id'].isin(df[df['infra_type'] == 'a_remplacer']['infra_id'])]
infra_detruites.plot(ax=axs, color='red', edgecolor='black', label='Infrastructures détruites')

#Afficher la légende
axs.legend()
```
